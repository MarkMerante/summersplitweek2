```python
import pandas as pd
import numpy as np
```

```python
week2 = pd.read_csv('summer_week2.csv')
week2.head()
```

```python
#filter to only LCS and LEC
temp = week2[week2.league.isin(['LEC','LCS'])]
NA_EU = temp[temp.position.isin(['team'])]
NA_EU
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
earlygame = NA_EU[['result','csdiffat15','xpdiffat15','golddiffat15','firstblood']]

plt.figure(figsize=(5,5))
sb.heatmap(earlygame.corr(),vmin=-1, cmap='coolwarm', annot=True)
plt.show()
earlygame.corr()
```

```python
sb.pairplot(earlygame,hue='result',palette='bwr')
```

```python
import statsmodels.api as sm
earlygame['intercept']=1.0
ind_cols=['csdiffat15','xpdiffat15','golddiffat15','firstblood']
logit = sm.Logit(earlygame['result'],earlygame[ind_cols])
result=logit.fit()
```

```python
print(result.summary())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = earlygame[['csdiffat15', 'xpdiffat15', 'golddiffat15','firstblood']]
y = earlygame['result']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
